```python
import torch
import numpy as np
from torch.autograd import Variable
from tqdm import tqdm_notebook as tqdm
class MatrixFactorization(torch.nn.Module):


  def __init__(self, no_Customers, no_items, n_factors=10):
      super().__init__()
      self.customer_factors=torch.nn.Embedding(no_Customers, n_factors) #create customer the embendings
      self.item_factors = torch.nn.Embedding(no_items, n_factors) #create item embeddings
      self.customer_factors.weight.data.uniform_(0, 0.05) #setting the weights to 0.05
      self.item_factors.weight.data.uniform_(0, 0.05)

  def  info(self, data):
      customers, items = data[:,0], data[:,1]
      return (self.customer_factors(customers)*self.item_factors(items)).sum(1)

  def predict(self, customers, item):
        return self.forward(customers, item)       
```

```python
# Creating the dataloader (necessary for PyTorch)
from torch.utils.data.dataset import Dataset
from torch.utils.data import DataLoader # package that helps transform your data to machine learning readiness


class Loader(Dataset):
    def __init__(self):
        self.rating= insertDf.copy()

        # Extract all customer IDs and iphone IDs
        customers = insertDf.Customer_Id.unique()
        iphones = insertDf.iphoneID.unique()
        
         #Producing new continuous IDs for customers and iphones/items
        
        # Unique values : index
        self.customerid2idx = {o:i for i,o in enumerate(customers)}
        self.iphoneid2idx = {o:i for i,o in enumerate(iphones)}

        # Obtained continuous ID for users and movies
        self.idx2customerid = {i:o for o,i in self.customerid2idx.items()}
        self.idx2iphoneid = {i:o for o,i in self.iphoneid2idx.items()}

        #return the ID from the indexed values as noted in the lambda function 
        self.x = self.rating.drop(['Ratings'], axis=1).values
        self.y = self.rating['iphoneID'].values
        self.x, self.y = torch.tensor(self.x), torch.tensor(self.y) # Transforms the data to tensors (ready for torch models.)

    def getIphone(self, index):
            return (self.x[index], self.y[index])
    def len(self):
            return len(self.rating)   
```

```python
num_epochs = 10
cuda = torch.cuda.is_available()
print("Is running on GPU:", cuda)

model = MatrixFactorization(no_Customers, no_items,n_factors=5)
print(model)

for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)
# GPU enable if you have a GPU..

if cuda:
    model = model.cuda()

#MSE loss
loss_fn = torch.nn.MSELoss()
# ADAM optimizier
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Train data so it can be readable in pytorch 
train_set = Loader()

train_loader = DataLoader(train_set, 10)
```

```python
for it in tqdm(range(num_epochs)):
    losses = []
for x, y in train_loader:   #y is ratings and x is the curstomer id and item id
         if cuda:
            x, y = x.cuda(), y.cuda()
            optimizer.zero_grad()
            outputs = model(x)
            loss = loss_fn(outputs.squeeze(), y.type(torch.float32))
            losses.append(loss.item())
            loss.backward()
            optimizer.step()
            
         print("iter #{}".format(it), "Loss:", sum(losses) / len(losses))
```

```python
# By training the model, we will have tuned latent factors for iPhones and customers.
c = 0
uw = 0
iw = 0 
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)
        if c == 0:
          uw = param.data
          c +=1
        else:
          iw = param.data
        #print('param_data', param_data)
```

```python
trained_phone_embeddings = model.item_factors.weight.data.cpu().numpy() #converting our item factors to numpy array
```

```python
len(trained_phone_embeddings) # unique item factor weights
```

```python
from sklearn.cluster import KMeans
# Fit the clusters based on the item weights
kmeans = KMeans(n_clusters=9, random_state=0).fit(trained_phone_embeddings)
```

```python
# 
for cluster in range(9):
  print("Cluster #{}".format(cluster))
  phone = []
  for iphoneidx in np.where(kmeans.labels_ == cluster)[0]:
    phoneid = train_set.idx2iphoneid[iphoneidx]
    rat_count = insertDf.loc[insertDf['iphoneID']==phoneid].count()[0]
    phone.append((iphoneName [phoneid], rat_count))
  for ipho in sorted(phone, key=lambda tup: tup[1], reverse=True)[:14]:
    print("\t", ipho[0])
```
